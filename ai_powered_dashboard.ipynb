```python
import pandas as pd

orders = pd.read_csv('/content/olist_orders_dataset.csv')
items = pd.read_csv('/content/olist_order_items_dataset.csv')
products = pd.read_csv('/content/olist_products_dataset.csv')
customers = pd.read_csv('/content/olist_customers_dataset.csv')

for name, df in [('orders', orders), ('items', items), ('products', products), ('customers', customers)]:
    print(f"\n{'='*40}\n{name.upper()} — shape: {df.shape}")
    print(df.columns.tolist())
    print(df.head(2))



# Load the translation file too
category_translation = pd.read_csv('/content/product_category_name_translation.csv')

# Merge: items + orders (on order_id) + products (on product_id) + customers (on customer_id) + category translation
df_merged = items.merge(orders, on='order_id', how='left')
df_merged = df_merged.merge(products, on='product_id', how='left')
df_merged = df_merged.merge(customers, on='customer_id', how='left')
df_merged = df_merged.merge(category_translation, on='product_category_name', how='left')

print("Merged shape:", df_merged.shape)
print("\nColumns:", df_merged.columns.tolist())
print("\nMissing values:")
print(df_merged.isnull().sum())
df_merged.head()
```

```python
def clean_olist_data(df):
    df = df.copy()

    # 1. Convert all date columns from string to datetime
    date_cols = ['order_purchase_timestamp', 'order_approved_at',
                 'order_delivered_carrier_date', 'order_delivered_customer_date',
                 'order_estimated_delivery_date', 'shipping_limit_date']
    for col in date_cols:
        df[col] = pd.to_datetime(df[col], errors='coerce')

    # 2. Decision: only keep 'delivered' orders for business KPIs
    # Cancelled/unavailable/processing orders don't represent real completed revenue,
    # and would distort metrics like delivery time (since they may have null delivery dates).
    # This is a deliberate business-logic decision, not just "cleaning" — worth explaining in interviews.
    before = len(df)
    df = df[df['order_status'] == 'delivered'].copy()
    print(f"Filtered to 'delivered' orders only: {before} -> {len(df)} rows ({before - len(df)} removed)")

    # 3. Handle missing product category (1,603 rows / ~1.4% of data)
    # Rather than dropping these (losing real revenue data), we label them explicitly.
    # This preserves the row for revenue/KPI totals while being honest that category is unknown.
    df['product_category_name_english'] = df['product_category_name_english'].fillna('unknown_category')

    # 4. Drop rows still missing delivery date after filtering to 'delivered' status
    # (a small edge case: status says delivered but date wasn't recorded — a genuine data quality issue)
    before = len(df)
    df = df.dropna(subset=['order_delivered_customer_date'])
    print(f"Dropped rows with missing delivery date despite 'delivered' status: {before - len(df)} removed")

    # 5. Feature engineering: actual delivery time in days
    df['delivery_days'] = (df['order_delivered_customer_date'] - df['order_purchase_timestamp']).dt.days

    # 6. Feature engineering: delivery performance vs estimate (negative = early, positive = late)
    df['delivery_vs_estimate_days'] = (df['order_delivered_customer_date'] - df['order_estimated_delivery_date']).dt.days

    # 7. Sanity checks
    invalid_delivery = (df['delivery_days'] < 0).sum()
    print(f"Rows with negative delivery time (data errors): {invalid_delivery}")

    invalid_price = (df['price'] < 0).sum()
    print(f"Rows with negative price (should be 0): {invalid_price}")

    # 8. Freight ratio — the "discount equivalent" insight for this dataset
    # High freight relative to price is a real margin/logistics problem, same role as
    # Superstore's discount-vs-profit relationship.
    df['freight_pct_of_price'] = round((df['freight_value'] / df['price']) * 100, 2)

    return df

df_clean = clean_olist_data(df_merged)
print("\nCleaned shape:", df_clean.shape)
print("\nRemaining missing values:")
print(df_clean[['product_category_name_english', 'delivery_days', 'delivery_vs_estimate_days', 'freight_pct_of_price']].isnull().sum())
df_clean.head()
```

```python
def calculate_olist_kpis(df):
    kpis = {}

    # --- Core financial metrics ---
    kpis['total_sales'] = round(df['price'].sum(), 2)
    kpis['total_freight'] = round(df['freight_value'].sum(), 2)
    kpis['total_orders'] = df['order_id'].nunique()
    kpis['avg_order_value'] = round(kpis['total_sales'] / kpis['total_orders'], 2)
    kpis['avg_freight_pct_of_price'] = round(df['freight_pct_of_price'].mean(), 2)

    # --- Delivery performance ---
    kpis['avg_delivery_days'] = round(df['delivery_days'].mean(), 2)
    kpis['avg_delivery_vs_estimate_days'] = round(df['delivery_vs_estimate_days'].mean(), 2)
    pct_late = (df['delivery_vs_estimate_days'] > 0).mean() * 100
    kpis['pct_orders_delivered_late'] = round(pct_late, 2)

    # --- Category breakdown ---
    category_sales = df.groupby('product_category_name_english')['price'].sum().sort_values(ascending=False)
    kpis['top_category'] = category_sales.index[0]
    kpis['top_category_sales'] = round(category_sales.iloc[0], 2)
    kpis['category_breakdown_top10'] = category_sales.head(10).round(2).to_dict()

    # --- State (region) breakdown ---
    state_sales = df.groupby('customer_state')['price'].sum().sort_values(ascending=False)
    kpis['top_state'] = state_sales.index[0]
    kpis['state_breakdown_top10'] = state_sales.head(10).round(2).to_dict()

    # --- Freight burden by category (interesting insight: which categories suffer worst freight ratios) ---
    freight_by_category = df.groupby('product_category_name_english')['freight_pct_of_price'].mean().sort_values(ascending=False)
    kpis['highest_freight_burden_category'] = freight_by_category.index[0]
    kpis['highest_freight_burden_pct'] = round(freight_by_category.iloc[0], 2)
    kpis['lowest_freight_burden_category'] = freight_by_category.index[-1]
    kpis['lowest_freight_burden_pct'] = round(freight_by_category.iloc[-1], 2)

    # --- High freight burden impact (equivalent to Superstore's "high discount" analysis) ---
    high_freight = df[df['freight_pct_of_price'] > 50]  # freight costs more than half the item price
    kpis['pct_orders_high_freight_burden'] = round((len(high_freight) / len(df)) * 100, 2)

    # --- Monthly trend ---
    df['Order Month'] = df['order_purchase_timestamp'].dt.to_period('M').astype(str)
    monthly_sales = df.groupby('Order Month')['price'].sum().round(2)
    kpis['monthly_sales_trend'] = monthly_sales.to_dict()

    return kpis, df

kpis, df_clean = calculate_olist_kpis(df_clean)

import json
print(json.dumps(kpis, indent=2, default=str))
```

```python
import plotly.express as px
import plotly.graph_objects as go

def create_monthly_trend_chart(kpis):
    months = list(kpis['monthly_sales_trend'].keys())
    sales = list(kpis['monthly_sales_trend'].values())
    fig = px.line(x=months, y=sales, title="Monthly Sales Trend",
                  labels={'x': 'Month', 'y': 'Sales (R$)'}, markers=True)
    fig.update_layout(xaxis_tickangle=-45, height=400)
    return fig

def create_category_breakdown_chart(kpis):
    categories = list(kpis['category_breakdown_top10'].keys())
    sales = list(kpis['category_breakdown_top10'].values())
    fig = px.bar(x=categories, y=sales, title="Top 10 Categories by Sales",
                 labels={'x': 'Category', 'y': 'Sales (R$)'}, color=categories)
    fig.update_layout(height=450, showlegend=False, xaxis_tickangle=-30)
    return fig

def create_state_breakdown_chart(kpis):
    states = list(kpis['state_breakdown_top10'].keys())
    sales = list(kpis['state_breakdown_top10'].values())
    fig = px.bar(x=states, y=sales, title="Top 10 States by Sales",
                 labels={'x': 'State', 'y': 'Sales (R$)'}, color=states)
    fig.update_layout(height=400, showlegend=False)
    return fig

def create_freight_burden_chart(df):
    freight_by_cat = df.groupby('product_category_name_english')['freight_pct_of_price'].mean().sort_values(ascending=False).head(15)
    fig = px.bar(x=freight_by_cat.values, y=freight_by_cat.index, orientation='h',
                 title="Freight Cost as % of Price — Top 15 Worst Categories",
                 labels={'x': 'Freight % of Price', 'y': 'Category'})
    fig.add_vline(x=50, line_dash="dash", line_color="red")  # 50% threshold marker
    fig.update_layout(height=500)
    return fig
```

```python
import numpy as np

def detect_anomalies_olist(kpis, threshold=2.0, min_sales_cutoff=5000):
    # Exclude months with unrealistically low sales (data collection ramp-up, not real anomalies)
    filtered_trend = {
        month: sales for month, sales in kpis['monthly_sales_trend'].items()
        if sales >= min_sales_cutoff
    }

    months = list(filtered_trend.keys())
    sales = np.array(list(filtered_trend.values()))

    mean_sales = sales.mean()
    std_sales = sales.std()
    z_scores = (sales - mean_sales) / std_sales

    anomalies = []
    for month, sale, z in zip(months, sales, z_scores):
        if abs(z) > threshold:
            anomalies.append({
                'month': month,
                'sales': round(float(sale), 2),
                'z_score': round(float(z), 2),
                'direction': 'spike' if z > 0 else 'drop'
            })

    excluded_months = [m for m in kpis['monthly_sales_trend'] if m not in filtered_trend]

    return anomalies, mean_sales, std_sales, excluded_months

anomalies, mean_sales, std_sales, excluded_months = detect_anomalies_olist(kpis, threshold=2.0, min_sales_cutoff=100000)

print(f"Excluded months (incomplete/ramp-up data): {excluded_months}")
print(f"\nMean monthly sales (filtered): R${mean_sales:,.2f}")
print(f"Std deviation: R${std_sales:,.2f}")
print(f"\nAnomalies detected (|z-score| > 2.0):")
for a in anomalies:
    print(f"  {a['month']}: R${a['sales']:,.2f} (z={a['z_score']}, {a['direction']})")

if not anomalies:
    print("  None found at threshold 2.0")
```

```python
def create_anomaly_chart_olist(kpis, anomalies, excluded_months):
    filtered_trend = {k: v for k, v in kpis['monthly_sales_trend'].items() if k not in excluded_months}
    months = list(filtered_trend.keys())
    sales = list(filtered_trend.values())

    fig = go.Figure()
    fig.add_trace(go.Scatter(x=months, y=sales, mode='lines+markers',
                              name='Monthly Sales', line=dict(color='#636efa')))

    if anomalies:
        anomaly_months = [a['month'] for a in anomalies]
        anomaly_sales = [a['sales'] for a in anomalies]
        fig.add_trace(go.Scatter(x=anomaly_months, y=anomaly_sales, mode='markers',
                                  name='Anomaly', marker=dict(color='red', size=14, symbol='x')))

    fig.update_layout(title="Monthly Sales Trend with Anomaly Detection (Incomplete Months Excluded)",
                       xaxis_title="Month", yaxis_title="Sales (R$)",
                       xaxis_tickangle=-45, height=450)
    return fig

fig5 = create_anomaly_chart_olist(kpis, anomalies, excluded_months)
fig5.show()
```

```python
def generate_executive_summary_olist(kpis, anomalies):
    context = {
        "total_sales": f"R${kpis['total_sales']:,.2f}",
        "total_orders": f"{kpis['total_orders']:,}",
        "avg_order_value": f"R${kpis['avg_order_value']:,.2f}",
        "avg_freight_pct_of_price": f"{kpis['avg_freight_pct_of_price']}%",
        "avg_delivery_days": kpis['avg_delivery_days'],
        "avg_delivery_vs_estimate_days": kpis['avg_delivery_vs_estimate_days'],
        "pct_orders_delivered_late": f"{kpis['pct_orders_delivered_late']}%",
        "top_category": kpis['top_category'],
        "top_state": kpis['top_state'],
        "highest_freight_burden_category": kpis['highest_freight_burden_category'],
        "highest_freight_burden_pct": f"{kpis['highest_freight_burden_pct']}%",
        "lowest_freight_burden_category": kpis['lowest_freight_burden_category'],
        "lowest_freight_burden_pct": f"{kpis['lowest_freight_burden_pct']}%",
        "pct_orders_high_freight_burden": f"{kpis['pct_orders_high_freight_burden']}%",
        "detected_anomalies": [
            {**a, 'sales': f"R${a['sales']:,.2f}"} for a in anomalies
        ]
    }

    prompt = f"""You are a senior business analyst writing an executive summary for a Brazilian e-commerce marketplace's leadership team.

You must ONLY use the numbers provided below, exactly as formatted. Do NOT invent, estimate, or assume any figure not explicitly given.

DATA:
{json.dumps(context, indent=2)}

Write a concise executive summary (250-300 words) with these sections:
1. **Overview** — headline performance in 2-3 sentences
2. **Key Insight** — the single most important finding (hint: look closely at freight cost as a % of price, and how it varies by category)
3. **Delivery Performance** — comment on delivery speed vs estimates
4. **Anomaly Note** — briefly mention the detected anomaly, if any
5. **Recommendations** — exactly 3 specific, actionable business recommendations

Write in a professional but direct tone, as if presenting to a VP of Operations. Do not use generic filler language."""

    model = genai.GenerativeModel('gemini-flash-latest')
    response = model.generate_content(prompt)
    return response.text

executive_summary = generate_executive_summary_olist(kpis, anomalies)
print(executive_summary)
```

```python
from reportlab.lib.pagesizes import letter
from reportlab.lib.units import inch
from reportlab.lib import colors
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle, Image
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from datetime import datetime
import re

def clean_markdown_for_pdf(text):
    text = re.sub(r'\*\*(.*?)\*\*', r'<b>\1</b>', text)
    text = re.sub(r'^### (.*?)$', r'<b><font size=13>\1</font></b>', text, flags=re.MULTILINE)
    return text

def generate_pdf_report_olist(kpis, anomalies, executive_summary, output_path="business_report.pdf"):
    doc = SimpleDocTemplate(output_path, pagesize=letter,
                             topMargin=0.6*inch, bottomMargin=0.6*inch)
    styles = getSampleStyleSheet()
    story = []

    title_style = ParagraphStyle('Title', parent=styles['Title'], fontSize=20, spaceAfter=6)
    story.append(Paragraph("E-Commerce Business Insights Report", title_style))
    story.append(Paragraph(f"Generated {datetime.now().strftime('%B %d, %Y')}", styles['Normal']))
    story.append(Spacer(1, 20))

    story.append(Paragraph("Key Performance Indicators", styles['Heading2']))
    kpi_data = [
        ["Metric", "Value"],
        ["Total Sales", f"R${kpis['total_sales']:,.2f}"],
        ["Total Orders", f"{kpis['total_orders']:,}"],
        ["Avg Order Value", f"R${kpis['avg_order_value']:,.2f}"],
        ["Avg Freight % of Price", f"{kpis['avg_freight_pct_of_price']}%"],
        ["Avg Delivery Days", f"{kpis['avg_delivery_days']}"],
        ["Avg Delivery vs Estimate", f"{kpis['avg_delivery_vs_estimate_days']} days"],
        ["% Orders Delivered Late", f"{kpis['pct_orders_delivered_late']}%"],
        ["Top Category", kpis['top_category']],
        ["Top State", kpis['top_state']],
        ["Highest Freight Burden Category", f"{kpis['highest_freight_burden_category']} ({kpis['highest_freight_burden_pct']}%)"],
    ]
    kpi_table = Table(kpi_data, colWidths=[3*inch, 3*inch])
    kpi_table.setStyle(TableStyle([
        ('BACKGROUND', (0,0), (-1,0), colors.HexColor('#636efa')),
        ('TEXTCOLOR', (0,0), (-1,0), colors.white),
        ('FONTNAME', (0,0), (-1,0), 'Helvetica-Bold'),
        ('FONTSIZE', (0,0), (-1,-1), 10),
        ('GRID', (0,0), (-1,-1), 0.5, colors.grey),
        ('ROWBACKGROUNDS', (0,1), (-1,-1), [colors.white, colors.HexColor('#f0f0f0')]),
        ('TOPPADDING', (0,0), (-1,-1), 6),
        ('BOTTOMPADDING', (0,0), (-1,-1), 6),
    ]))
    story.append(kpi_table)
    story.append(Spacer(1, 20))

    story.append(Paragraph("Executive Summary", styles['Heading2']))
    summary_html = clean_markdown_for_pdf(executive_summary)
    for para in summary_html.split('\n\n'):
        if para.strip():
            story.append(Paragraph(para.strip(), styles['Normal']))
            story.append(Spacer(1, 8))

    story.append(Spacer(1, 10))
    footer_style = ParagraphStyle('Footer', parent=styles['Normal'], fontSize=8, textColor=colors.grey)
    story.append(Paragraph(
        "This report was generated using an AI-augmented analytics pipeline. "
        "All figures are computed directly from source data; narrative summary is AI-generated "
        "and grounded strictly in the computed metrics above.", footer_style
    ))

    doc.build(story)
    print(f"PDF report saved to: {output_path}")
    return output_path

# Build a filename like: business_report_2026-07-23.pdf
today_str = datetime.now().strftime('%Y-%m-%d')
output_filename = f"business_report_{today_str}.pdf"

pdf_path = generate_pdf_report_olist(kpis, anomalies, executive_summary, output_path=output_filename)
```

```python
import gradio as gr
import tempfile
import os

def process_dashboard_olist(csv_file):
    if csv_file is None:
        return None, None, None, None, "Please upload a CSV file.", None, None

    # Load (already cleaned/merged CSV)
    df = pd.read_csv(csv_file.name, parse_dates=[
        'order_purchase_timestamp', 'order_approved_at',
        'order_delivered_carrier_date', 'order_delivered_customer_date',
        'order_estimated_delivery_date', 'shipping_limit_date'
    ])

    # KPIs
    kpi_results, df = calculate_olist_kpis(df)

    # Charts
    chart1 = create_monthly_trend_chart(kpi_results)
    chart2 = create_category_breakdown_chart(kpi_results)
    chart3 = create_state_breakdown_chart(kpi_results)
    chart4 = create_freight_burden_chart(df)

    # Anomaly detection
    anomaly_results, _, _, _ = detect_anomalies_olist(kpi_results, threshold=2.0, min_sales_cutoff=100000)

    # AI Summary
    summary_text = generate_executive_summary_olist(kpi_results, anomaly_results)

    # PDF
    today_str = datetime.now().strftime('%Y-%m-%d')
    temp_pdf_path = os.path.join(tempfile.gettempdir(), f"business_report_{today_str}.pdf")
    generate_pdf_report_olist(kpi_results, anomaly_results, summary_text, temp_pdf_path)

    kpi_markdown = f"""
### Key Metrics
| Metric | Value |
|---|---|
| Total Sales | R${kpi_results['total_sales']:,.2f} |
| Total Orders | {kpi_results['total_orders']:,} |
| Avg Order Value | R${kpi_results['avg_order_value']:,.2f} |
| Avg Freight % of Price | {kpi_results['avg_freight_pct_of_price']}% |
| Avg Delivery Days | {kpi_results['avg_delivery_days']} |
| % Delivered Late | {kpi_results['pct_orders_delivered_late']}% |
| Top Category | {kpi_results['top_category']} |
| Top State | {kpi_results['top_state']} |
"""

    return chart1, chart2, chart3, chart4, kpi_markdown, summary_text, temp_pdf_path

with gr.Blocks(title="AI Powered E-Commerce Insights Dashboard") as demo:
    gr.Markdown("# 📦 AI-Powered E-Commerce Insights Dashboard")
    gr.Markdown("Upload the cleaned Olist merged CSV to generate KPIs, charts, an AI executive summary, and a downloadable PDF report.")

    with gr.Row():
        file_input = gr.File(label="Upload Merged CSV", file_types=[".csv"])
        submit_btn = gr.Button("Generate Dashboard", variant="primary")

    kpi_output = gr.Markdown(label="KPIs")

    with gr.Row():
        chart1_output = gr.Plot(label="Monthly Trend")
        chart2_output = gr.Plot(label="Category Breakdown")

    with gr.Row():
        chart3_output = gr.Plot(label="State Breakdown")
        chart4_output = gr.Plot(label="Freight Burden by Category")

    summary_output = gr.Textbox(label="AI Executive Summary", lines=15)
    pdf_output = gr.File(label="Download PDF Report")

    submit_btn.click(
        fn=process_dashboard_olist,
        inputs=[file_input],
        outputs=[chart1_output, chart2_output, chart3_output, chart4_output, kpi_output, summary_output, pdf_output]
    )

demo.launch(share=True, debug=True)
```
